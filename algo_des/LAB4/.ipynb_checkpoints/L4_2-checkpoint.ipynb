```python
import networkx as nx
import numpy as np

g_size = 12
g = nx.Graph()
V = np.arange(g_size)
E = list()
V_pos = list()

# losowanie pozycji z przedziałów [0, 100], [0, 100]
for i in V:
    V_pos.append(np.random.randint(0, 100, 2))

# losowanie krawędzi
for v in V:
    for i in range(2):
        k = np.random.randint(0, len(V))
        if k != v:
            E.append([v, k])

g.add_nodes_from(V)
nx.draw(g, pos=V_pos)
```

Przykład dla samodzielnie wylosowanego grafu bez krawędzi

```python
# funkcja dzieli graf na dwa mniejsze po połowie wierzchołków
# parametr num_of_r wyznacza ile razy ma wywołać się rekurencyjnie
def split_graph(grph, num_of_r):
    nodes = list(grph.nodes)
    A = nx.Graph()
    a_nodes = nodes[0: int(len(nodes)/2)]
    print('a_nodes:', a_nodes)
    A.add_nodes_from(a_nodes)
    # losowe dodawanie krawędzi w obrębie nowo powstałego grafu
    for n in range(3*len(a_nodes)):
        A.add_edge(np.random.randint(0, len(a_nodes)), np.random.randint(0, len(a_nodes)))

    B = nx.Graph()
    b_nodes = nodes[int(len(nodes)/2):]
    # print('b_nodes', b_nodes)
    # losowe dodawanie krawędzi w obrębie nowo powstałego grafu
    B.add_nodes_from(b_nodes)

    if num_of_r != 1:
        return split_graph(A, num_of_r-1), split_graph(B, num_of_r-1)
    else:
        return A, B


A, B = split_graph(g, num_of_r=2)

Ac, Ad = A[0], A[1]
Bc, Bd = B[0], B[1]

color_map = []
for node in g.nodes:
    if node in Ac:
        color_map.append('pink')
    elif node in Ad:
        color_map.append('purple')
    elif node in Bc:
        color_map.append('orange')
    elif node in Bd:
        color_map.append('yellow')


nx.draw(g, pos=V_pos, node_color=color_map,with_labels=True)
nx.draw_networkx_edges(Ac, pos=V_pos)
nx.draw_networkx_edges(Ac, pos=V_pos[int(g_size/4):int(g_size/2)])
nx.draw_networkx_edges(Ac, pos=V_pos[int(g_size/2):int(3*g_size/4)])
nx.draw_networkx_edges(Ac, pos=V_pos[int(3*g_size/4):g_size])
```

Zastosowanie funkcji o grafie wylosowanym funkcją random_internet_graph z pakietu networkx.

Jak narazie program dzieli tylko wierzchołki najpierw na 2 grupy (A, B) a potem każdą grupę jeszcze na pół (C, D).

```python
G = nx.random_internet_as_graph(n=60, seed=33)


A, B = split_graph(G, num_of_r=2)

Ac, Ad = A[0], A[1]
Bc, Bd = B[0], B[1]

color_map = []
for node in G.nodes:
    if node in Ac:
        color_map.append('pink')
    elif node in Ad:
        color_map.append('purple')
    elif node in Bc:
        color_map.append('orange')
    elif node in Bd:
        color_map.append('yellow')

nx.draw(G, node_color=color_map,with_labels=True)
```
